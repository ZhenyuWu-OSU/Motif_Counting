```python
import re
import pandas as pd
import numpy as np
import os

```

```python
## make function
def countmotif(inputstr:str,pattern:str):
    i=0
    matchtime=0
    for i in range(0,len(inputstr)):
        test_str=inputstr[i:i+3]
        #print(test_str)
        matchtime=matchtime+(test_str==pattern)
    return matchtime

```

```python
path = "./HumanChr-FASTA/" #文件夹目录
files= os.listdir(path) #得到文件夹下的所有文件名称
s = []
for file in files: #遍历文件夹
        print(file)
     #if not os.path.isdir(file): #判断是否是文件夹，不是文件夹才打开
        fr = open(path+"/"+file) #打开文件
        dict={}
        for line in fr:
            if line.startswith('>'):
                name1=line.split("OX=")[0]
                #print(name1)
                name = name1[1:150]
                #print(name)
                dict[name]=''
            else:
                dict[name]+=line.replace('\n','')
        fr.close()
        ##motifcount_result=pd.DataFrame(dict)
        ## motifcount_result

        my_df = pd.DataFrame.from_dict(dict, orient='index')
        my_df

        testmotif=["GPP","DPP","PPP","SPP","PPG","PPA","PPV","PPE",
                   "PPD","APP","MPP","DSP","DDP","PDP","DNP","DGP",
                   "PDK","PDG","GDP","PDI","PGP","PDR","DPG","RDK",
                   "PDA","PDV","DVG","DDG","GGT"]
        for name in testmotif:
            namecount=[]
            for ID in dict.keys():
                namecount.append(countmotif(dict[ID],name))   
            my_df[str(name)]=namecount

        length=[]
        for ID in dict.keys():
            length.append(len(dict[ID]))

        length

        my_df["length"]=length
        my_df.to_csv(file+".csv")
```
